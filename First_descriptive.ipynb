# Getting Into Stats

```python
print('Mode:\n',Inc_Exp_Data_df.mode())
```

```python
# calculate mean, median, and standard deviation for a column of the data
mean = np.mean(Inc_Exp_Data_df['Mthly_HH_Income'])
median = np.median(Inc_Exp_Data_df['Mthly_HH_Income'])
std_dev = np.std(Inc_Exp_Data_df['Mthly_HH_Income'])

# calculate minimum and maximum values
minimum = np.min(Inc_Exp_Data_df['Mthly_HH_Income'])
maximum = np.max(Inc_Exp_Data_df['Mthly_HH_Income'])

# calculate range
range_value = np.ptp(Inc_Exp_Data_df['Mthly_HH_Income'])

# calculate percentiles
percentile_25 = np.percentile(Inc_Exp_Data_df['Mthly_HH_Income'], 25)
percentile_75 = np.percentile(Inc_Exp_Data_df['Mthly_HH_Income'], 75)

# calculate interquartile range
iqr = percentile_75 - percentile_25

# calculate skewness
skewness = Inc_Exp_Data_df['Mthly_HH_Income'].skew()

# calculate kurtosis
kurtosis = Inc_Exp_Data_df['Mthly_HH_Income'].kurt()


# add titles and labels to the plot
#plt.title('Histogram of Column Name')
#plt.xlabel('Value')
#plt.ylabel('Frequency')

# print out the results
#print('Basic Statistics:\n', stats)
print('Mean:', mean)
print('Median:', median)
print('Standard Deviation:', std_dev)
print('Minimum:', minimum)
print('Maximum:', maximum)
print('Range:', range_value)
print('25th Percentile:', percentile_25)
print('75th Percentile:', percentile_75)
print('Interquartile Range:', iqr)
print('Skewness:', skewness)
print('Kurtosis:', kurtosis)
```

# Geometric Mean

# Doughnut Diagram

```python
grouped_data = Inc_Exp_Data_df.groupby('Highest_Qualified_Member').sum()
# plot the chart
plt.pie(grouped_data['Mthly_HH_Income'], labels=grouped_data.index, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.title('Monthly Household Income By Highest graduate')

# create a circle in the center of the chart to make it a doughnut chart
circle = plt.Circle((0,0), 0.7, color='white')
fig = plt.gcf()
fig.gca().add_artist(circle)

plt.show()
```
